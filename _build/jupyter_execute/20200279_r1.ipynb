##### 1.1) Derivaci칩n 
siendo: 
$$T=YT$$
y, $$Yd= Y-T$$ 
por ende, $$ Yd=(1 - t)Y $$
Entonces, al reemplazar valores, en la funci칩n de demanda de consumo resultar칤a:
$$ C = C_0 + bYd  $$

$$ C = Co + b(1-t)Y $$
##### 1.2) Explicaci칩n 
La funci칩n de Demanda de Consumo parte de que las familias toman decisiones de gasto y ahorro. Asimismo, la suma de salarios de los trabajadores y de los beneficios de los propietarios del stock de capital es igual al Ingreso Agregado (Y) y como cierta parte de este ingreso (Y) se entrega al Estado como tributaci칩n (T), con lo que queda se decide consumir y ahorrar.

Adem치s, sus valores significan:
$C_0$ es el nivel b치sico de consumo, es decir, el consumo aut칩nomo, y $b$ representa la propensi칩n marginal a consumir, en otras palabras, representa el cambio en el consumo por unidad de cambio en el ingreso disponible. Esta se encuentra entre cero y uno, 0 < 洧녪 < 1. 

##### 1.3) R칠plica de gr치fico 

```python
# Par치metros

Y_size = 100 

Co = 35
b = 0.8
t = 0.3

Y = np.arange(Y_size)

# Ecuaci칩n de la curva del ingreso de equilibrio

def C(Co, b, t, Y):
    C = Co + b*(1-t)*Y
    return C

C = C(Co, b, t, Y)
```

```python
# Gr치fico

# Dimensiones del gr치fico
y_max = np.max(Y)
fig, ax = plt.subplots(figsize=(10, 8))

custom_xlim = (0, 130)
custom_ylim = (20, 130)

plt.setp(ax, xlim=custom_xlim, ylim=custom_ylim)


# Curvas a graficar
ax.plot(Y, C, label = "Consumo", color = "C3") #Demanda agregada

# Eliminar las cantidades de los ejes
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

# Texto agregado
    # punto de equilibrio
plt.text(80, 95, '$PMgC = 郊쓖/郊쓫^d = b = 0.5$', fontsize = 13, color = 'black')


# T칤tulo y leyenda
ax.set(title="Funci칩n de demanda de Consumo", xlabel= '$Y^d$', ylabel= 'C')
ax.legend() #mostrar leyenda

plt.show()
```

##### 2.1) Explicaci칩n 

Los gastos de inversi칩n aumentan el stock de capital y la producci칩n de largo plazo de la econom칤a.
La funci칩n de Demanda de Inversi칩n $$I= I_0-hr$$ los representa. Aqu칤, $I_0$ es la inversi칩n aut칩noma,$h$ es la sensibilidad de un inversionista ante la tasa de inter칠s real y $r$ es la tasa de inter칠s real. 

Los valores de $r$ y $h$ se encuentran en un rango de 0 a 1.Entonces, hay una variaci칩n inversa entre la tasa de inter칠s real $r$ y la funci칩n de demanda de Inversi칩n $I$ y una variaci칩n directa entre la

##### 2.2)Gr치fico

```python
#Par치metros
Io = 40
h = 0.7

r = np.arange(start = 0.01, stop = 1, step =0.01)  
r

# Ecuaci칩n de la curva del ingreso de equilibrio

def I(Io, h, r):
    I = Io - h*r
    return I

I = I(Io, h, r)
```

$$ DA = (C_0 + bYd + I_0 + G_0 + X_0 - hr) - mYd $$

Luego de reemplazar ecuaciones y realizar algunas operaciones algebraicas se obtiene:

$$ DA = (C_0 + b(1-t)Y + I_0 + G_0 + X_0 - hr) - m(1-t)Y $$

$$ DA = (C_0 + I_0 + G_0 + X_0 - hr) + (b - m)(1 - t)Y $$

$$ DA = 풤_0 + 풤_1Y $$

$$ Y = (C_0 + I_0 + G_0 + X_0 - hr)+(b - m)*(1 - t)Y $$

$$ Y =  \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

##### b) Analice una Pol칤tica fiscal expansiva con una reducci칩n de la Tasa de Tributaci칩n ($풊t < 0$). El an치lisis debe ser intuitivo, matem치tico y gr치fico.
- An치lisis intuitivo: 쮺ontradicci칩n?
$$ t  Co  DA  DA > Y  Y $$
$$ t  M  DA  DA < Y  Y $$

- An치lisis matem치tico: $갡t < 0    갡Y?$

```python
Co, Io, Go, Xo, h, r, b, m, t = symbols('Co Io Go Xo h r b m t')

f = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))

df_t = diff(f, t)
df_t #갡Y/갡t
```

Considernado el diferencial de $갡t$:

$$ \frac{갡Y}{갡t} = \frac{(m-b)(Co + Go + Io + Xo - hr)}{(1-(1-t)(b-m)+1)^2} $$
y:
- Sabiendo que b > m, entonces $(m-b) < 0$
- Los componentes aut칩nomos no cambian: $갡C_0 = 갡I_0 = 갡X_0 = 갡h = 갡r = 0$
- Cualquier n칰mero elevado al cuadrado ser치 positivo: $ (1-(1-t)(b-m)+1)^2 > 0 $

Entonces:

$$ \frac{갡Y}{갡t} = \frac{(-)}{(+)} $$

Dado que $갡t < 0$, la divisi칩n de un n칰mero positivo y otro nu칰mero positivo da otro positivo:

$$ \frac{갡Y}{(-)} = \frac{(-)}{(+)} $$

$$ 갡Y = \frac{(-)(-)}{(+)} $$
